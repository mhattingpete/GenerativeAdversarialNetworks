# Conditional DCGAN for MNIST

## Imports

```python
import torch
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "1"

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from viz import updatable_display2

import torch.nn as nn
import torch.autograd as autograd
import torch.nn.functional as F
from torch import nn, optim
from torchvision import transforms, datasets
from torch.utils.data import DataLoader

from generators import CondConvGenerator
from discriminators import CondConvDiscriminator
from utils import img2vec,vec2img,sample_noise,true_target,fake_target,onehot,num_parameters

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(device)
```

## Dataset loading

```python
image_size = 28

compose = transforms.Compose([transforms.ToTensor(),transforms.Normalize((.5, .5, .5), (.5, .5, .5))])
dataset = datasets.MNIST(root='MNIST',train=True,transform=compose,download=True)
```

### WGAN-GP: gradients penalty

```python
LAMBDA = 10 # recommended value from the original paper

def calc_gradient_penalty(discriminator,real_data,real_label,fake_data,fake_label):    
    batch_size = real_data.size(0)
    alpha = torch.rand(batch_size,1)
    alpha = alpha.expand(real_data.size())
    alpha = alpha.to(device)
    alpha_label = torch.rand(batch_size,1)
    alpha_label = alpha_label.expand(real_label.size())
    alpha_label = alpha_label.to(device)
    
    interpolates = alpha * real_data + ((1 - alpha) * fake_data)
    interpolates = autograd.Variable(interpolates,requires_grad=True)
    interpolates = interpolates.to(device)
    
    interpolates_label = alpha_label * real_label + ((1 - alpha_label) * fake_label)
    interpolates_label = autograd.Variable(interpolates_label,requires_grad=True)
    interpolates_label = interpolates_label.to(device)
    
    disc_interpolates = discriminator(vec2img(interpolates,image_size),vec2img(interpolates_label,image_size)).view(-1)
        
    gradients = autograd.grad(outputs=disc_interpolates,inputs=(interpolates,interpolates_label),
                              grad_outputs=torch.ones(disc_interpolates.size()).to(device),
                              create_graph=True,retain_graph=True,only_inputs=True)
    gradient_penalty = ((gradients[0].norm(p=2,dim=1)-1)**2).mean()*LAMBDA
    gradient_penalty += ((gradients[1].norm(p=2,dim=1)-1)**2).mean()*LAMBDA
    return gradient_penalty
```

## Training setup

```python
batch_size = 128
#m = int(np.sqrt(batch_size))-1 # the sqrt of number of test samples
lr = 1e-3
dropout_prob = 0.2
noise_dim = 100
conditional_dim = 10
output_size = 1
clipping_value = 1e-2

num_test_samples = conditional_dim**2
test_noise = sample_noise(conditional_dim,noise_dim,device).repeat(conditional_dim,1)[:,:,None,None]
test_labels = torch.zeros(conditional_dim, 1)
for i in range(conditional_dim-1):
    temp = torch.ones(conditional_dim,1)+i
    test_labels = torch.cat([test_labels,temp],0)
test_labels_onehot = onehot(test_labels.type(torch.LongTensor),conditional_dim).type(torch.FloatTensor).to(device)

# label preprocess
fill = torch.zeros([conditional_dim,conditional_dim,image_size,image_size])
for i in range(conditional_dim):
    fill[i,i,:,:] = 1

# intialize models
generator = CondConvGenerator(input_size=noise_dim,conditional_size=conditional_dim,hidden_size=32,output_size=output_size).to(device)
discriminator = CondConvDiscriminator(input_size=output_size,conditional_size=conditional_dim,hidden_size=32,output_size=output_size).to(device)

# otpimizers
g_optimizer = optim.Adam(generator.parameters(),lr=lr)
d_optimizer = optim.Adam(discriminator.parameters(),lr=lr)

# data loader
data_loader = DataLoader(dataset,batch_size=batch_size,shuffle=True)

# Create logger instance
dis = updatable_display2(['train'],["epoch","d_error","g_error"])
# Total number of epochs to train
num_epochs = 100
global_step = 0
epoch = 0
d_error = 0
g_error = 0

print("# of parameters for G: \t {}".format(num_parameters(generator)))
print("# of parameters for D: \t {}".format(num_parameters(discriminator)))
```

### Train Generator

```python
def train_generator(noise,fake_label,optimizer):
    '''
    Train the generator to generate realistic samples and thereby fool the discriminator
    '''
    # Reset gradients
    optimizer.zero_grad()
    
    # Sample noise and generate fake data
    prediction = discriminator(noise,fake_label)
    loss = - torch.mean(prediction)
    loss.backward()
    optimizer.step()
    return loss
```

### Train Discriminator

```python
def train_discriminator(real_data,real_label,fake_data,fake_label,optimizer):
    '''
    Train the discriminator to distinguish between real and fake data
    '''
    # Reset gradients
    optimizer.zero_grad()

    # 1.1 Train on Real Data
    prediction_real = discriminator(real_data,real_label)

    # 1.2 Train on Fake Data
    prediction_fake = discriminator(fake_data,fake_label)
    
    # gradients penalty (Lipschitz condition) (WGAN-GP)
    batch_size = real_data.size(0)
    penalty = calc_gradient_penalty(discriminator,real_data.view(batch_size,-1),real_label.view(batch_size,-1),
                                    fake_data.view(batch_size,-1),fake_label.view(batch_size,-1))
    loss = prediction_fake.mean() - prediction_real.mean() + penalty
    loss.backward()
    
    # 1.3 Update weights with gradients
    optimizer.step()
    
    # 1.4 clip D parameters (Lipschitz condition) (orignal WGAN)
    #for p in discriminator.parameters():
    #    p.data.clamp_(-clipping_value, clipping_value)
    
    return loss
```

## Train the model

```python
gen_steps = 1
gen_train_freq = 5
try:
    while epoch < num_epochs:
        for n_batch,(real_batch,label_batch) in enumerate(data_loader):
            N = real_batch.size(0)
            # 1. Train Discriminator
            real_data = real_batch.to(device)
            real_label_onehot = onehot(label_batch[:,None],conditional_dim).float().to(device)
            real_label_image = fill[label_batch].to(device)
            # Generate fake data and detach 
            # (so gradients are not calculated for generator)
            noise_tensor = sample_noise(N,noise_dim,device)[:,:,None,None]
            fake_label = torch.randint(low=0,high=conditional_dim,size=(N,1)).squeeze().type(torch.LongTensor)
            label_onehot = onehot(fake_label,conditional_dim).type(torch.FloatTensor).to(device)
            fake_label_image = fill[fake_label].type(torch.FloatTensor).to(device)
            with torch.no_grad():
                fake_data = generator(noise_tensor,label_onehot[:,:,None,None]).detach()
            # Train D
            d_error = train_discriminator(real_data,real_label_image,fake_data,fake_label_image,d_optimizer)

            # 2. Train Generator every 'gen_train_freq' steps
            if global_step % gen_train_freq == 0:
                for _ in range(gen_steps):
                    # Generate fake data
                    fake_label = torch.randint(low=0,high=conditional_dim,size=(N,1)).squeeze().type(torch.LongTensor)
                    label_onehot = onehot(fake_label,conditional_dim).type(torch.FloatTensor).to(device)
                    fake_label_image = fill[fake_label].type(torch.FloatTensor).to(device)
                    fake_data = generator(sample_noise(N,noise_dim,device)[:,:,None,None],label_onehot[:,:,None,None])
                    # Train G
                    g_error = train_generator(fake_data,fake_label_image,g_optimizer)
                    g_error = g_error.item()

            # Log batch error and delete tensors
            dis.update(global_step,'train',{"epoch":epoch,"d_error":d_error.item(),"g_error":g_error} )
            global_step += 1
            del fake_data
            del real_data
            del noise_tensor

            # Display Progress every few batches
            if global_step % 50 == 0:
                test_images = vec2img(generator(test_noise,test_labels_onehot[:,:,None,None]),image_size)
                test_images = test_images.data
                canvas = np.zeros((image_size*conditional_dim,image_size*conditional_dim))
                q = 0
                for i in range(conditional_dim):
                    for j in range(conditional_dim):
                        canvas[i*image_size:(i+1)*image_size,j*image_size:(j+1)*image_size] = test_images[q]
                        q += 1
                dis.display(scale=False)
                plt.figure(figsize=(10,10))
                plt.imshow(canvas,cmap='gray')
                plt.axis("off")
                if epoch % 50 == 0:
                    plt.savefig("Figures/GAN-MNIST-Epoch="+str(epoch)+".png")
                plt.show()
        epoch += 1
except:
    test_images = vec2img(generator(test_noise,test_labels_onehot[:,:,None,None]),image_size)
    test_images = test_images.data
    canvas = np.zeros((image_size*conditional_dim,image_size*conditional_dim))
    q = 0
    for i in range(conditional_dim):
        for j in range(conditional_dim):
            canvas[i*image_size:(i+1)*image_size,j*image_size:(j+1)*image_size] = test_images[q]
            q+=1
    dis.display(scale=False)
    plt.figure(figsize=(10,10))
    plt.imshow(canvas,cmap='gray')
    plt.axis("off")
    plt.savefig("Figures/DCGAN-MNIST.png")
    plt.show()
```
